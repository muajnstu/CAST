```python
import pandas as pd
import io
from google.colab import files
```

```python

# Step 1: Upload the raw dataset
print("Please upload your dataset for categorization:")
uploaded = files.upload()
```

```python
# Step 2: Load the dataset into a DataFrame
file_name = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[file_name]))
```

```python
# Step 3: Define categorization function for research reproducibility
# Logic: 0 -> 0 (No Gap), 1 to 2 -> 1 (Low Gap), >2 -> 2 (High Gap)
def categorize_gap(value):
    try:
        # Convert to float to ensure decimals are handled correctly
        val = float(value)

        if val == 0:
            return 0
        elif 0 < val <= 2:
            return 1
        elif val > 2:
            return 2
        else:
            return 0  # Default case for unexpected values
    except (ValueError, TypeError):
        # Assign 0 if the cell contains non-numeric strings (e.g., "Minor Gap")
        return 0
```

```python

# Step 4: Apply the transformation across the entire dataset
# applymap ensures the function is applied to every single cell
processed_df = df.applymap(categorize_gap)
```

```python
# Step 5: Displaying the first few rows for validation
print("\nTransformation complete. Review the first 5 rows below:")
print(processed_df.head())
```

```python
# Step 6: Export and Download with the specific filename
output_filename = 'Gap categorized.csv'
processed_df.to_csv(output_filename, index=False)
files.download(output_filename)

print(f"\nFinal file '{output_filename}'")
```
