```python
import pandas as pd
import io
from google.colab import files
```

```python
# Step 1: Upload the raw dataset
print("Please upload your Likert scale dataset for categorization:")
uploaded = files.upload()
```

```python
# Step 2: Load the dataset into a DataFrame
file_name = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[file_name]))
```

```python
# Step 3: Define categorization function based on updated criteria
# Logic:
# Strongly Agree/Agree (1, 2) -> 0
# Neutral (3) -> 1
# Disagree/Strongly Disagree (4, 5) -> 2
def categorize_likert_scale(value):
    try:
        val = float(value)

        # Applying the specific scale mapping
        if val == 1 or val == 2:
            return 0  # Agree/Strongly Agree
        elif val == 3:
            return 1  # Neutral
        elif val == 4 or val == 5:
            return 2  # Disagree/Strongly Disagree
        else:
            return 1  # Defaulting outliers to Neutral (or adjust as needed)

    except (ValueError, TypeError):
        # Assign 1 (Neutral) if the cell contains non-numeric strings
        return 1
```

```python
# Step 4: apply the transformation across the entire dataset
# This replaces every cell value with the new category (0, 1, 2)
processed_df = df.applymap(categorize_likert_scale)
```

```python
# Step 5: Displaying the first few rows for validation
print("\nTransformation complete. Review the mapping below:")
print(processed_df.head())
```

```python
# Step 6: Export and Download with the requested filename
output_filename = 'Gap categorized.csv'
processed_df.to_csv(output_filename, index=False)
files.download(output_filename)

print(f"\nFinal file '{output_filename}' is ready for submission.")
```
